```python
from matplotlib import pyplot as plt
import polars as pl
import seaborn as sns
import japanize_matplotlib
import numpy as np
```

# PU Learning

```python
supervised = np.load("./sup_bin_valid_accuracies.npy")[:200]
pu = np.load("./pu_valid_accuracies.npy")[:200]

plt.plot(supervised, label="Supervised")
plt.plot(pu, label="PU")
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.legend()
plt.title("PU分類と教師あり学習の比較")
plt.ylim(0.5, 1.0)
plt.show()
```
